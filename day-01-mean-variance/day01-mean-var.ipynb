```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
```

```python
data1=np.random.normal(loc=50,scale=2,size=100) #low variance
data2=np.random.normal(loc=50,scale=10,size=100) #medium variance
data3=np.random.normal(loc=50,scale=25,size=100) #high variance
```

```python
data1
```

```python
data2
```

```python
data3
```

```python
data3_with_outliers=np.append(data3,[200,-100])
```

```python
data3_with_outliers
```

```python
np.mean(data3), np.mean(data3_with_outliers)
```

```python
def describe(data,name):
    print(name)
    print(f"mean:{np.mean(data):.3f}")
    print(f"var:{np.var(data):.3f}")
    print(f"stdev:{np.std(data):.3f}")
    print("-"*30)
dataset=[(data1,"low variance"),(data2,"med variance"),(data3,"high variance"),(data3_with_outliers,"outlier+high variance")]
for data,name in dataset:
    describe(data,name)
    
```

```python
plt.figure(figsize=(14,6))
plt.subplot(1,4,1)
plt.hist(data1)
plt.title("low var")
plt.subplot(1,4,2)
plt.hist(data2)
plt.title("med var")
plt.subplot(1,4,3)
plt.hist(data3)
plt.title("high var")
plt.subplot(1,4,4)
plt.hist(data3_with_outliers)
plt.title("high var+outlier")
plt.tight_layout()
plt.show()
```
